```python
from tkinter import *

#defining login function
def Info():
    
    #getting form data
    srollno=stu_rollno.get()
    sname=stu_name.get()
    saddress=stu_address.get()
    smobile= stu_mobile.get()
    sdepartment=stu_department.get()
    sfees=stu_fees.get()
    
    #applying empty validation
    if srollno=='' or sname=='' or saddress=='' or smobile=='' or sdepartment=='' or sfees=='':
        message.set("fill the empty field!!!")
    else:
        message.set("Student Information Registered Successfully!!!")
    
#defining loginform function
def information_form():
    global information_screen
    information_screen = Tk()
    #Setting title of screene
    information_screen.title("Student Entry Form")
    #setting height and width of screen
    information_screen.geometry("500x500")
    
    #declaring variable
    global  message;
    global stu_rollno
    global stu_name
    global stu_address
    global stu_mobile
    global stu_department
    global stu_fees
   
    stu_rollno=StringVar()
    stu_name=StringVar()
    stu_address=StringVar()
    stu_mobile=StringVar()
    stu_department=StringVar()
    stu_fees=StringVar()
    message=StringVar()
    
    #Creating layout of login form
    Label(information_screen,width="300", text="Please enter details below", bg="orange",fg="white").pack()
    
    #Labels All
    Label(information_screen, text="Rollno  * ").place(x=20,y=40)
    Label(information_screen, text="Name * ").place(x=20,y=80)
    Label(information_screen, text="Address * ").place(x=20,y=120)
    Label(information_screen, text="Mobile* ").place(x=20,y=160)
    Label(information_screen, text="Department * ").place(x=20,y=200)
    Label(information_screen, text="Fees * ").place(x=20,y=240)
    
    # Textboxes All
    Entry(information_screen, textvariable=stu_rollno).place(x=90,y=42)
    Entry(information_screen, textvariable=stu_name,show="*").place(x=90,y=82)
    Entry(information_screen, textvariable=stu_address).place(x=90,y=122)
    Entry(information_screen, textvariable=stu_mobile,show="*").place(x=90,y=162)
    Entry(information_screen, textvariable=stu_department).place(x=90,y=202)
    Entry(information_screen, textvariable=stu_fees,show="*").place(x=90,y=242)
    
    #Label for displaying login status[success/failed]
    Label(information_screen, text="",textvariable=message).place(x=95,y=265)
    
    #Login button
    Button(information_screen, text="Submit", width=10, height=1, bg="orange",command=Info).place(x=105,y=285)
    information_screen.mainloop()
    
#calling function Loginform
information_form()
```
